```python
import os
import json
from glob import glob
from pathlib import Path

import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from ultralytics import YOLO
from pycocotools.coco import COCO
from pycocotools.cocoeval import COCOeval
```

```python
SAMPLE_IMAGE_MAP = {
    "bright_high_contrast_blurry": "02a3ba4c3886fe9a_jpg.rf.3eed71c4a4d26606beff7edb40836fdd.jpg",
    "bright_high_contrast_sharp": "008e36a5a0623445_jpg.rf.c0df6282d2fe4484c4c6de82acbb2ab5.jpg",
    "bright_low_contrast_blurry": "55-D_NQ_NP_610094-MLM31234159132_062019-W_jpg.rf.609fce89ef92b6596eed1b9f76af96bc.jpg",
    "bright_low_contrast_sharp": "Amazon-com_-20SunplusTrade-20License-20Plate-20Frame_-202-20Pack_yythk_jpeg_jpg.rf.b63a624bab479dcc8e11b13fd9520836.jpg",
    "low_light_high_contrast_blurry": "CarLongPlateGen2695_jpg.rf.08da95274df6d79faece1813ae567f6f.jpg",
    "low_light_high_contrast_sharp": "Best-20Electric-20Car-20License-20Plates_jpg.rf.0fe0d3ef48931b9eefc2091ed85e0e28.jpg",
    "low_light_low_contrast_blurry": "CarLongPlateGen1365_jpg.rf.cd95ac1dce970c9d893de67187c5be9e.jpg"
}

# Convert YOLO labels â†’ COCO GT JSON
def yolo_to_coco_gt(images_dir, labels_dir, output_json, category_map):
    image_files = sorted(glob(f"{images_dir}/*.jpg") + glob(f"{images_dir}/*.png"))
    coco = {
        "images": [], "annotations": [],
        "categories": [{"id": cid, "name": name} for name,cid in category_map.items()]
    }
    ann_id = 0
    for img_idx, img_path in enumerate(image_files):
        img = cv2.imread(img_path)
        h, w = img.shape[:2]
        image_id = img_idx + 1
        coco["images"].append({
            "id": image_id,
            "file_name": Path(img_path).name,
            "height": h,
            "width": w
        })
        lbl_file = Path(labels_dir)/f"{Path(img_path).stem}.txt"
        if lbl_file.exists():
            for line in open(lbl_file):
                cls, xc,yc,bw,bh = map(float, line.split())
                x = (xc - bw/2)*w
                y = (yc - bh/2)*h
                bw *= w; bh *= h
                coco["annotations"].append({
                    "id": ann_id,
                    "image_id": image_id,
                    "category_id": int(cls),
                    "bbox": [x, y, bw, bh],
                    "area": bw*bh,
                    "iscrowd": 0
                })
                ann_id += 1

    with open(output_json, "w") as f:
        json.dump(coco, f)

# Run YOLO inference â†’ COCO Predictions JSON
def predict_with_yolo(model, images_dir, category_map, output_json):
    image_files = sorted(glob(f"{images_dir}/*.jpg") + glob(f"{images_dir}/*.png"))
    coco = {
        "images": [], "annotations": [],
        "categories": [{"id": cid, "name": name} for name,cid in category_map.items()]
    }
    ann_id = 0
    for img_idx, img_path in enumerate(image_files):
        img = cv2.imread(img_path)
        h, w = img.shape[:2]
        image_id = img_idx + 1
        coco["images"].append({
            "id": image_id,
            "file_name": Path(img_path).name,
            "height": h,
            "width": w
        })
        res = model(img_path)[0]
        for *box, conf, cls in res.boxes.data.tolist():
            x1,y1,x2,y2 = box
            bw, bh = x2-x1, y2-y1
            coco["annotations"].append({
                "id": ann_id,
                "image_id": image_id,
                "category_id": int(cls),
                "bbox": [x1, y1, bw, bh],
                "score": float(conf)
            })
            ann_id += 1

    with open(output_json, "w") as f:
        json.dump(coco, f)

# Evaluate with COCOeval (bbox mode) and return metrics dict
def coco_evaluate(gt_json, pred_json):
    coco_gt = COCO(gt_json)
    pred_data = json.load(open(pred_json))
    anns = pred_data["annotations"]
    coco_dt = coco_gt.loadRes(anns)

    coco_eval = COCOeval(coco_gt, coco_dt, iouType="bbox")
    coco_eval.params.useSegm = False
    coco_eval.evaluate()
    coco_eval.accumulate()
    coco_eval.summarize()

    stats = coco_eval.stats  # 12-element array
    keys = [
        "AP@[.50:.95]","AP@.50","AP@.75",
        "AP_small","AP_med","AP_large",
        "AR@[.50:.95]","AR@.50","AR@.75",
        "AR_small","AR_med","AR_large"
    ]
    return dict(zip(keys, stats.tolist()))

def visualize_sample(folder_name, images_dir, gt_annotations, dt_annotations, category_map, out_path):
    sample_img_name = SAMPLE_IMAGE_MAP.get(folder_name)
    if sample_img_name is None:
        print(f"âš ï¸ No sample image configured for {folder_name}, skipping visualization.")
        return

    img_entry = next(img for img in gt_annotations["images"] if img["file_name"] == sample_img_name)
    img_path = Path(images_dir)/sample_img_name
    img = cv2.cvtColor(cv2.imread(str(img_path)), cv2.COLOR_BGR2RGB)

    # draw GT (green)
    gt_img = img.copy()
    for ann in filter(lambda a: a["image_id"]==img_entry["id"], gt_annotations["annotations"]):
        x,y,w,h = ann["bbox"]
        cv2.rectangle(gt_img, (int(x),int(y)), (int(x+w),int(y+h)), (0,255,0), 2)

    # draw preds (red)
    dt_img = img.copy()
    for ann in filter(lambda a: a["image_id"]==img_entry["id"], dt_annotations["annotations"]):
        x,y,w,h = ann["bbox"]; s=ann["score"]
        cv2.rectangle(dt_img, (int(x),int(y)), (int(x+w),int(y+h)), (255,0,0), 2)
        cv2.putText(dt_img, f"{s:.2f}", (int(x),int(y)-5),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,0,0), 2)

    fig, ax = plt.subplots(1,2,figsize=(12,6))
    fig.suptitle(f"Folder: {folder_name}", fontsize=16, y=0.98)
    ax[0].imshow(gt_img); ax[0].set_title("GT (green)"); ax[0].axis("off")
    ax[1].imshow(dt_img); ax[1].set_title("Pred (red)"); ax[1].axis("off")
    fig.tight_layout(rect=[0, 0, 1, 0.94])

    fig.savefig(out_path)
    plt.show()
    plt.close(fig)


base = Path("archive/test_multidim")
model = YOLO("license_plate_detector-YOLOv8n.pt")
category_map = {"license_plate": 0}

os.makedirs("eval_results/plots", exist_ok=True)
records = []

for folder in sorted(base.iterdir()):
    if not folder.is_dir():
        continue
    name = folder.name
    imgs = folder/"images"
    lbls = folder/"labels"

    gt_json   = f"gt_{name}.json"
    pred_json = f"pred_{name}.json"

    print(f"â†’ Evaluating: {name}")

    yolo_to_coco_gt(str(imgs), str(lbls), gt_json, category_map)
    predict_with_yolo(model, str(imgs), category_map, pred_json)

    metrics = coco_evaluate(gt_json, pred_json)
    metrics["folder"] = name
    records.append(metrics)

    # visualize only the configured sample
    gt_data = json.load(open(gt_json))
    dt_data = json.load(open(pred_json))
    vis_path = f"eval_results/plots/{name}.png"
    visualize_sample(name, str(imgs), gt_data, dt_data, category_map, vis_path)

# write one CSV for all
df = pd.DataFrame(records)
df.to_csv("eval_results/tide_metrics_summary.csv", index=False)
print("âœ… Done! CSV â†’ eval_results/tide_metrics_summary.csv")

```
